```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
```

```python
data = pd.read_csv('../models/model_history/test5.csv')
```

```python
data1 = pd.read_csv('../models/model_history/First.txt', sep=",", header=None)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(data['val_loss'], label = 'Test Loss')
ax.plot(data['loss'], label = 'Training Loss')
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
ax.set_title('Model Loss')
ax.legend()
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(data['val_accuracy'], label = 'Test Loss')
ax.plot(data['accuracy'], label = 'Training Loss')
ax.set_xlabel('Epoch')
ax.set_ylabel('Accuracy')
ax.set_title('Model Accuracy')
ax.legend()
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(data['val_precision'], label = 'Test Precision')
ax.plot(data['precision'], label = 'Training Precision')
ax.plot(data['val_recall'], label = 'Test Recall')
ax.plot(data['recall'], label = 'Training Recall')
ax.set_xlabel('Epoch')
ax.set_ylabel('Rate')
ax.set_title('Model Precision & Recall')
ax.legend()
```
