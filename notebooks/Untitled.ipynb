```python
import numpy as np
import matplotlib.pyplot as plt
import glob
%matplotlib inline 

from skimage import io, color, filters, feature, restoration
from skimage.transform import rescale, resize, downscale_local_mean
from skimage.color import rgb2gray
from scipy.spatial.distance import squareform
from scipy.misc import imread
```

```python
lines = glob.glob('../data/output/train/lines/*')
lines
```

```python
def resize_images(image_list):
    resized_images = []
    for image in image_list:
        read = io.imread(image)
        resized = resize(read, (30, 30))
        resized_images.append(resized)
    return np.array(resized_images)
```

```python
resized_images = resize_images(lines)
```

```python
image = lines[0]
```

```python
image = io.imread(image)
```

```python
image_rescaled = rescale(image, 0.25, anti_aliasing=False)
```

```python
io.imshow(image_rescaled)
```

```python
io.imshow(image)
```

```python
image_resize = resize(image, (30, 30))
```

```python
io.imshow(image_resize)
```
